## Part 2 — Gaussian NLL: Predicting Mean and Variance

A standard regression model outputs a single value $\hat{y} = f(\mathbf{x})$.
A **probabilistic** regression model outputs a full distribution. The simplest
approach assumes a Gaussian:

$$Y | \mathbf{x} \sim \mathcal{N}\big(\hat{\mu}(\mathbf{x}),\; \hat{\sigma}^2(\mathbf{x})\big)$$

The model has **two outputs** per patient:
- $\hat{\mu}(\mathbf{x})$: the predicted mean (like Lab 7)
- $\hat{\sigma}^2(\mathbf{x})$: the predicted variance (NEW — patient-specific uncertainty)

The **Gaussian Negative Log-Likelihood (NLL)** loss trains both heads jointly:

$$\mathcal{L} = \frac{1}{2n} \sum_{i=1}^{n} \left[ \frac{(y_i - \hat{\mu}_i)^2}{\hat{\sigma}_i^2} + \log \hat{\sigma}_i^2 \right]$$

This loss has two competing terms:
- $\frac{(y - \hat{\mu})^2}{\hat{\sigma}^2}$: rewards predicting the correct mean, **down-weighted** when $\hat{\sigma}^2$ is large
- $\log \hat{\sigma}^2$: penalizes predicting large variance (prevents the model from cheating by making $\hat{\sigma} \to \infty$)

## Part 3 — Quantile Regression

Gaussian NLL assumes the errors are normally distributed. **Quantile regression**
makes *no distributional assumptions*. Instead of modeling $P(Y | X)$ as a Gaussian,
it directly estimates specific quantiles (e.g., the 10th, 50th, 90th percentiles)
using the **pinball loss**:

$$L_q(y, \hat{y}) = \begin{cases} q \cdot (y - \hat{y}) & \text{if } y \geq \hat{y} \\ (1-q) \cdot (\hat{y} - y) & \text{if } y < \hat{y} \end{cases}$$

Or equivalently: $L_q(y, \hat{y}) = q \cdot \max(y - \hat{y}, 0) + (1-q) \cdot \max(\hat{y} - y, 0)$

Intuition:
- $q = 0.5$: pinball loss = MAE/2 → predicts the median
- $q = 0.9$: under-prediction (actual above predicted) is penalized 9× more than over-prediction → model errs on the high side → predicts the 90th percentile
- $q = 0.1$: over-prediction is penalized 9× more → predicts the 10th percentile

## Part 4 — Calibration of Probabilistic Predictions

A probabilistic model is only useful if its uncertainty estimates are **calibrated**:
if the model says "there's a 90% chance the true value is in [a, b]", then ~90% of
the time the true value should actually fall in that interval.

Two key concepts:
- **Coverage**: the fraction of true values inside the predicted interval (should match the nominal level)

## Part 5 — Heteroscedastic vs. Homoscedastic Models

A **homoscedastic** model predicts the same uncertainty for all patients (one global

### Solution: Gaussian NLL (Part 2)

```python
def gaussian_nll_solution(y, mu_hat, sigma2_hat, eps=1e-8):
    """Gaussian Negative Log-Likelihood loss."""
    sigma2_safe = np.maximum(sigma2_hat, eps)
    squared_term = (y - mu_hat) ** 2 / sigma2_safe
    log_term = np.log(sigma2_safe)
    nll = (1 / (2 * len(y))) * np.sum(squared_term + log_term)
    return nll

# Verify
y_ck = np.array([10.0, 15.0, 20.0])
mu_ck = np.array([10.0, 15.0, 20.0])
sig2_ck = np.array([1.0, 1.0, 1.0])

print(f"Perfect mu, σ²=1:   NLL = {gaussian_nll_solution(y_ck, mu_ck, sig2_ck):.4f}  (≈ 0)")
print(f"Perfect mu, σ²=100: NLL = {gaussian_nll_solution(y_ck, mu_ck, np.array([100,100,100.])):.4f}  (≈ log(100)/2 = {np.log(100)/2:.2f})")

mu_bad = np.array([20.0, 5.0, 10.0])
print(f"Bad mu, σ²=1:       NLL = {gaussian_nll_solution(y_ck, mu_bad, sig2_ck):.4f}  (very high)")
print(f"Bad mu, σ²=100:     NLL = {gaussian_nll_solution(y_ck, mu_bad, np.array([100,100,100.])):.4f}  (lower — honest about uncertainty)")
```

### Solution: Two-Headed Linear Model (Part 2)

```python
# The key TODO lines in the training loop:
# mu_tr    = X_train @ w_mu + b_mu
# log_s_tr = X_train @ w_ls + b_ls
# d_mu     = (mu_tr - y_train) / sig2_tr
# d_log_s  = 0.5 * (1.0 - (y_train - mu_tr)**2 / sig2_tr)

# Here's a standalone solution:
def train_gaussian_nll_solution(X_tr, y_tr, X_vl, y_vl, lr_mu=0.001, lr_ls=0.0001, n_epochs=1500):
    np.random.seed(42)
    n, d_feat = X_tr.shape

    w_mu = np.random.randn(d_feat) * 0.01
    b_mu = y_tr.mean()
    w_ls = np.zeros(d_feat)
    b_ls = np.log(y_tr.var())

    train_nlls, val_nlls = [], []

    for epoch in range(n_epochs):
        mu_tr = X_tr @ w_mu + b_mu
        log_s_tr = X_tr @ w_ls + b_ls
        sig2_tr = np.exp(np.clip(log_s_tr, -10, 10))

        nll_tr = gaussian_nll_solution(y_tr, mu_tr, sig2_tr)

        d_mu = (mu_tr - y_tr) / sig2_tr
        d_log_s = 0.5 * (1.0 - (y_tr - mu_tr)**2 / sig2_tr)

        dw_mu = X_tr.T @ d_mu / n
        db_mu = d_mu.mean()
        dw_ls = X_tr.T @ d_log_s / n
        db_ls = d_log_s.mean()

        w_mu -= lr_mu * dw_mu;  b_mu -= lr_mu * db_mu
        w_ls -= lr_ls * dw_ls;  b_ls -= lr_ls * db_ls

        mu_vl = X_vl @ w_mu + b_mu
        sig2_vl = np.exp(np.clip(X_vl @ w_ls + b_ls, -10, 10))
        nll_vl = gaussian_nll_solution(y_vl, mu_vl, sig2_vl)

        train_nlls.append(nll_tr)
        val_nlls.append(nll_vl)

        if epoch % 300 == 0 or epoch == n_epochs - 1:
            print(f"Epoch {epoch:4d} | Train NLL: {nll_tr:.4f} | Val NLL: {nll_vl:.4f} | "
                  f"Mean σ(val): {np.sqrt(sig2_vl).mean():.2f}")

    return w_mu, b_mu, w_ls, b_ls, train_nlls, val_nlls

print("Solution function defined.")
print("Key insight: two separate 'heads' — one for mean (mu), one for log-variance (log_s).")
print("The exp() of log_s ensures variance is always positive.")
```

### Solution: Pinball Loss (Part 3)

```python
def pinball_loss_solution(y, y_hat, q):
    """Pinball (quantile) loss."""
    residual = y - y_hat
    loss = q * np.maximum(residual, 0) + (1 - q) * np.maximum(-residual, 0)
    return np.mean(loss)

# Verify
y_ck = np.array([10.0, 15.0, 20.0])
yhat_ck = np.array([12.0, 15.0, 18.0])

pb_50 = pinball_loss_solution(y_ck, yhat_ck, q=0.5)
mae_half = np.mean(np.abs(y_ck - yhat_ck)) / 2
print(f"q=0.5: {pb_50:.4f}, MAE/2: {mae_half:.4f}  (should match)")

pb_90 = pinball_loss_solution(y_ck, yhat_ck, q=0.9)
pb_10 = pinball_loss_solution(y_ck, yhat_ck, q=0.1)
print(f"q=0.9: {pb_90:.4f}  (under-predictions costly)")
print(f"q=0.1: {pb_10:.4f}  (over-predictions costly)")
```

### Solution: Empirical Coverage (Part 4)

```python
def empirical_coverage_solution(y_true, lower, upper):
    """Fraction of true values inside [lower, upper]."""
    inside = (y_true >= lower) & (y_true <= upper)
    return np.mean(inside)

# Verify
y_demo = np.array([1, 2, 3, 4, 5.0])
lo_demo = np.array([0.5, 1.5, 2.5, 3.5, 4.5])
hi_demo = np.array([1.5, 2.5, 3.5, 4.5, 5.5])
print(f"Coverage: {empirical_coverage_solution(y_demo, lo_demo, hi_demo):.1%}  (should be 100%)")

hi_narrow = np.array([1.5, 2.5, 3.5, 4.5, 4.9])  # last one misses
print(f"Coverage: {empirical_coverage_solution(y_demo, lo_demo, hi_narrow):.1%}  (should be 80%)")
```

3. Heteroscedasticity (larger residuals for larger tumors) means the model is unreliably
   precise for large tumors. A point estimate of "25 mm" with true uncertainty of ±5 mm is
   very different from ±0.5 mm. Clinically, this matters for surgical planning, radiation
   targeting, and treatment response monitoring — all of which depend on knowing how much
   to trust the prediction.

**Reflection 2.1 — The Gaussian NLL Loss**

1. Predicting $\log \sigma^2$ ensures positivity via the $\exp()$ transform. If we predicted
   $\sigma^2$ directly, the model could output negative values, which are meaningless. The
   log-parameterization maps $\mathbb{R} \to \mathbb{R}^+$ naturally.

2. If $\hat{\sigma}^2 \to 0$, the $(y - \hat{\mu})^2 / \hat{\sigma}^2$ term explodes for
   any imperfect prediction, driving the loss to infinity. The model can only achieve low
   loss by claiming low uncertainty where it actually *is* accurate. This self-regulating
   property forces the model to be calibrated: it learns to be uncertain where it's
   inaccurate.

3. If $\hat{\sigma}^2 = c$ (constant), the NLL becomes
   $\frac{1}{2n}\sum[(y_i - \hat{\mu}_i)^2/c + \log c]$. The $\log c$ term is constant
   w.r.t. $\hat{\mu}$, so minimizing NLL is equivalent to minimizing
   $\sum(y_i - \hat{\mu}_i)^2$ = MSE. Thus MSE is the homoscedastic special case of
   Gaussian NLL.

**Reflection 3.1 — Quantile vs. Gaussian**

1. Quantile regression is better for skewed targets (e.g., hospital LOS, medical costs)
   because it naturally produces asymmetric intervals. Gaussian NLL is simpler and works
   well when errors are approximately symmetric. For tumor radius (roughly symmetric after
   log-transform), Gaussian may suffice; for raw medical costs (heavily skewed), quantile
   regression is preferred.

2. Crossing quantiles are a fundamental limitation — separate models for each quantile
   don't coordinate. Fixes include: (a) sorting predicted quantiles post-hoc (simple but
   ad hoc), (b) using a single model that outputs all quantiles simultaneously with
   monotonicity constraints, (c) using a distributional model (like Gaussian NLL) that
   generates quantiles from a parametric distribution (which can't cross by construction).

3. Intervals should be wider for patients in data-sparse regions (unusual feature
   combinations) and for patients similar to those with high residuals in training. Features
   driving uncertainty might include rare diagnoses, extreme lab values, or demographic
   groups underrepresented in training data.

**Reflection 4.1 — Coverage vs. Sharpness**

1. Model B (85% coverage, 5 mm width) is likely more clinically useful because narrow
   intervals are actionable. But it needs calibration improvement — you could widen
   intervals slightly (e.g., multiply by a calibration factor) to reach 90% coverage.
   Model A's 20 mm intervals are so wide they're uninformative — "the tumor is somewhere
   between 5 and 25 mm" doesn't help a surgeon.

2. Coverage alone rewards trivially wide intervals. A useful prediction interval must be
   both *calibrated* (correct coverage) AND *sharp* (narrow). The ideal model is the one
   with the narrowest intervals that still maintain the nominal coverage level.

3. If the Gaussian model is over-confident (coverage < nominal), it may be because the
   true error distribution has heavier tails than a Gaussian — extreme errors are more
   common than the model expects. Under-confidence (coverage > nominal) means the model
   is hedging too much, which wastes clinical utility.

**Reflection 5.1 — Types of Uncertainty**

1. The Gaussian NLL model captures **aleatoric** uncertainty — it learns that measurement
   noise varies across patients (heteroscedasticity). It does NOT capture **epistemic**
   uncertainty — uncertainty from limited data. Epistemic uncertainty can be estimated
   via: (a) ensembles (train multiple models, look at disagreement), (b) MC Dropout
   (run the model multiple times with dropout, measure variance), (c) Bayesian approaches.

2. Yes, distinguishing them is crucial. High aleatoric uncertainty → "this patient is
   inherently unpredictable, order more tests." High epistemic uncertainty → "we haven't
   seen patients like this, be cautious and consider referring to a specialist." The
   clinical response is completely different.

3. If higher predicted uncertainty corresponds to higher actual error, the model's
   uncertainty estimates are *informative* and *useful* — they correctly identify which
   patients it's least confident about. This is the goal. If there's no relationship,
   the model's variance head is not well-calibrated, and the intervals shouldn't be
   trusted.

**Final Reflection Answers**

1. **Type of uncertainty:** A model with constant-width intervals is homoscedastic — it
   captures average aleatoric uncertainty but not patient-specific variance. It misses
   both heteroscedastic aleatoric uncertainty (some patients are inherently more variable)
   and epistemic uncertainty (the model hasn't seen similar patients).

2. **Clinical impact:** The deteriorating patient has much higher inherent variability
   (and the model should know less about where their creatinine is heading). A constant
   interval gives false reassurance — the clinician might not escalate monitoring for a
   patient whose true uncertainty should trigger urgent attention.

3. **How to fix it:** Start with Gaussian NLL (to learn heteroscedastic variance) — this
   is the simplest upgrade from MSE and directly addresses the constant-width problem.
   Then consider quantile regression if distributions are asymmetric. Add ensembling last
   for epistemic uncertainty. This ordering goes from simplest to most complex, each
   adding a new type of uncertainty quantification.

4. **Connection to Lab 7:** Lab 7 could only diagnose heteroscedasticity *after* seeing
   outcomes (in residual plots). Lab 9 *predicts* patient-specific uncertainty *before*
   the outcome is known. This is transformative: a clinician can now see "I'm 90% confident
   this patient's creatinine will be in [X, Y]" and use that to decide monitoring intensity,
   not just learn retrospectively that their predictions were unreliable for certain patients.
